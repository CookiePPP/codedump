```python
example_text_split = example_text.split("\n")
```

```python
new_line = ""
for i, line in enumerate(example_text_split):
    if ("-->" in line):
#        start_time = line.split(" ")[0]
        end_time = line.split(" ")[2]
        end_time = f"<{end_time}>"
    if ("<c>" in line):
        new_line += line.replace("</c>","").replace("<c>","") + end_time
print(new_line)
words = [x.split("<") for x in new_line.split(">") if x.strip()]
times = [x[1].split(":") for x in words]
words = [x[0] for x in words]
time_secs = [((float(hour)*3600) + (float(minute)*60) + float(second)) for hour, minute, second in times] # convert hours:minutes:seconds to seconds

segments = [""]
previous_finish_time = segment_start = time_secs[0]
print(f"{'Word':<10}|{'FinishTime':<10}|{'Seperation'}")
for word, finish_time in zip(words, time_secs):
    sep_time = round(finish_time-previous_finish_time,2)
    print(f"{word:<10}|{finish_time:<10}|{sep_time}")
    if (sep_time > 0.7) or (finish_time-segment_start > 8):
        segments[-1] = segments[-1].strip()
        segments.append(str(word))
        segment_start = finish_time
    elif sep_time > 0.65:
        segments[-1]+=f". {word.title()}".replace("  "," ")
    elif sep_time > 0.55:
        segments[-1]+=f", {word}".replace("  "," ")
    else:
        segments[-1]+=f" {word}".replace("  "," ")
    previous_finish_time = finish_time
print("\n".join(segments))
```
